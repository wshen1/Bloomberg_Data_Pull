```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_lst = ['https://esgnews.com/', 
           'https://www.esgtoday.com/',
          #'https://www.esgdive.com/'
          # 'https://www.reuters.com/sustainability/esg-investors/',
          # 'https://sustainabilitymag.com/esg'
          ]

pulled_news_lst = []
for url in url_lst:

   # url = "https://esgnews.com/"
    headers = {"User-Agent": "Mozilla/5.0"}
    
    r = requests.get(url, headers=headers)
    r.raise_for_status()
    
    soup = BeautifulSoup(r.text, "lxml")
    
    # Example: pull all article titles and links
    
    content_lst = []
    title_lst = []
    
    for a in soup.select("article h2 a"):   # change selector to fit the site
        title = a.get_text(strip=True)
        link = a.get("href")
        print(title, link)
    
        r1 = requests.get(link, headers=headers)
        r1.raise_for_status()
        
        soup = BeautifulSoup(r1.text, "lxml")
        
        # Extract title
        title = soup.select_one("h1").get_text(strip=True)
        
        # Extract publish date
        date = soup.select_one("time").get_text(strip=True) if soup.select_one("time") else None
        
        # Extract article text
        paragraphs = [p.get_text(strip=True) for p in soup.select("div.td-post-content p")]
        if not paragraphs:  # fallback if selector changes
            paragraphs = [p.get_text(strip=True) for p in soup.find_all("p")]
        
        content = "\n\n".join(paragraphs)
        title_lst.append(title )
        content_lst.append( paragraphs)

    
    Pulled_News = pd.DataFrame({
    "Title": title_lst,
    "Content": content_lst
    })
    Pulled_News['Source'] = url

    pulled_news_lst.append(Pulled_News)

Total_Pulled_News = pd.concat(pulled_news_lst).reset_index(drop = True)
Total_Pulled_News.to_csv('Pulled_Data/Pulled_News_20250918.csv', index = False)
```
